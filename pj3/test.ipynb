```python
import os
```

```python
img_dir = './dataset/test/rgb/'

rgb_img_paths = []
thermal_img_paths = []
for filename in os.listdir(img_dir):
    if not filename.endswith('.jpg'):
        continue
    rgb_file_path = os.path.join(img_dir, filename)
    thermal_file_path = rgb_file_path.replace("rgb/", "tir/").replace(".jpg", "R.jpg")
    
    rgb_img_paths.append(rgb_file_path)
    thermal_img_paths.append(thermal_file_path)
    
img_paths = list(zip(rgb_img_paths, thermal_img_paths))

nSamples = len(img_paths)
```
